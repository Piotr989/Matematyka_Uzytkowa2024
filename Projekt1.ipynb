# 2. Wyprowadzenie równania ruchu na niskich orbitach i uproszczenia

**Siły działające**

---

Przy niewielkich orbitach efekty związane z przyciąganiem Księżyca, bądź ciśnieniem słonecznym są pomijalnie małe [4] , dlatego przyjmiemy, że jedynymi siłami działającymi na orbitujący obiekt są: siła przyciągania ziemskiego i siła oporu atmosfery

$$
  F_{drag} + F_{grav} = F_{total} = ma, \tag{1}
$$

gdzie $m$ to jego masa, a $a$ to wartość przyspieszenia działająca na orbitujący obiekt. $F_{grav}$ to wartość siły grawitacji wynikającej z przyciągania Ziemi opisana wzorem

$$
  F_{grav} = G\frac{Mm}{R^2},
$$

gdzie $M$ to masa Ziemii, $G$ to stała grawitacyjna, a $R$ jest odległością obiektu na orbicie od środka Ziemii. $F_{drag}$ to wartość siły oporu atmofery wyrażana wzorem [5]

$$
  F_{drag} = \frac{1}{2}\rho CAv^2,
$$

przy czym $C$ to współczynnik oporu areodynamicznego, $\rho$ to gęstość ośrodka, $A$ to przekrój poprzeczny, a $v$ to wartość prędkości poruszania się.

**Wstępne założenia i równanie ruchu**

---

Badanie zachowania obiektów o dowolnie skomplikowanej geometrii jest bardzo trudne w praktyce [6], dlatego założymy, że obiekt jest kulą. Prędkość poruszania się jest tak wielka, że przyjmiemy współczynnik oporu $C$ za stały. Warto zwrócić uwagę, że dla obiektu o nietrywialnym kształcie poddanemu szybkim rotacjom można przybliżać przekrój poprzeczny średnią.  Co więcej, naturalną konsekwencją ogranicznia się tylko do sił oporu i grawitacji są orbity w całości zawarte w jednej płaszczyźnie. Przyjmijmy zatem środek Ziemii za początek układu współrzędnych, wtedy trajektoria ruchu po orbicie będzie funkcją $x:[0,T]\to \mathbb{R}^2$ spełniającą równanie różniczkowe Newtona

$$
 m\ddot x(t) = -F_{grav}\frac{x(t)}{|x(t)|} - F_{drag}\frac{\dot x(t)}{|\dot x(t)|}.
$$

Faktycznie jeśli $x(t) = (x_1(t), x_2(t))$ jest położeniem odłamka w chwili $t$, to $-x(t)$ jest w kierunku środka Ziemii, natomiast $\dot x(t) = \tfrac{\partial}{\partial t}x(t)$ jest wektorem prędkości zaczepionym w chwili $t$ w punkcie $x(t)$. Zauważmy jednocześnie, że długość wektora $x(t)$ jest jednocześnie odległością od środka Ziemii, zatem $R = |x(t)|$. Po podstawieniu $(1)$ otrzymujemy

$$
 m\ddot x(t) = -G\frac{Mm}{|x(t)|^2}\frac{x(t)}{|x(t)|} - \frac{1}{2}\rho CA|\dot x(t)|^2\frac{\dot x(t)}{|\dot x(t)|},
$$

$$
m\ddot x(t) = -G\frac{Mmx(t)}{|x(t)|^3} - \frac{1}{2}\rho CA|\dot x(t)|\dot x(t). \tag{2}
$$

Oczywiście gęstość atmosfery różni się w zależności od wysokości, zatem $\rho = \rho(h)$. Przyjmiemy więc, że atmosfera jest jednorodna, a jej gęstość zmniejsza się wykładniczo [7]

$$
  \rho(h) = \rho(|x(t)|) = \rho_0e^{-\tfrac{|x(t)| - 6378000}{H}},
$$

gdzie $\rho_0 = 1,225 \tfrac{kg}{m^3}$ to gęstość atmosfery na poziomie morza, natomiast $H = 7000[m]$.

**Implementacja i testy**

---

Do rozwiązywania równania $(2)$ skorzystamy z popularnego solvera ```solve_ivp``` z pakietu ```scipy.integrate```. Żeby jednak wprowadzić to równanie do programu sprowadzimy je najpierw do równania pierwszego rzędu. W tym celu wprowadzimy układ
$$
\begin{cases}
  y_1(t) = x(t), \newline
  y_2(t) = \dot x(t),
\end{cases}
$$

dzięki któremu równanie $(2)$ możemy zapisać jako układ równań pierwszego rzędu

$$
\begin{cases}
  \dot y_1(t) = y_2(t), \newline
  \dot y_2(t) = -G\frac{Mmy_1(t)}{|y_1(t)|^3} - \frac{1}{2}\rho(|y_1(t)|)CA|y_2(t)|y_2(t).
\end{cases} \tag{3}
$$

Definiując funkcję $F(t) = (y_1(t), y_2(t))$ otrzymujemy równanie wektorowe typu

$$
  \dot F(t) = Λ(F(t), t),
$$

gdzie $Λ$ jest funkcją opisującą układ $(3)$. Z uwagi na symetrię Ziemi założymy bez straty ogólności, że ruch zaczyna się w zenicie na wysokości $h_0$ i przebiega w kierunku ruchu wskazówek zegara z prędkością $ v_0 $. Wobec czego pełne zagadanienie Cauchy'ego wygląda następująco

$$
\begin{cases}
  \dot F(t) = \Lambda(F(t), t), \newline
  F(0) = ((0, h_0), \overrightarrow{[v_0, 0]}).
\end{cases}
$$


# Kod funkcji obliczającej trajektorię spadającego obiektu

Poniżej zaprezentowano kod funkcji ```simulate_trajectory()```, która służy do obliczania trajektorii obiektu na podstawie przedstawionych równań, z uwzględnieniem zadanych parametrów początkowych. Parametr ```initial_velocity``` jest listą zawierającą dwie wartości: prędkość poziomą i pionową. Ustawienie parametru ```make_picture``` na ```True``` spowoduje wygenerowanie wykresu trajektorii, natomiast przy jednoczesnym ustawieniu ```fall_info=True```, otrzymamy wykres wysokości nad powierzchnią Ziemii podczas ruchu. Funkcja zwraca tablicę ```[fall, num_orbits, max_distance, last_time_above_earth]```. Zmienna ```fall``` wskazuje, czy obiekt spadł na Ziemię w zadanym czasie. Zmienna ```num_orbits``` reprezentuje liczbę pełnych okrążeń Ziemi przed spadkiem, ```max_distance``` to maksymalna odległość od środka Ziemi osiągnięta podczas spadku, a ```last_time_above_earth``` to ostatni moment, w którym obiekt znajdował się nad Ziemią. Należy zauważyć, że rozwiązanie jest obliczane dla całego zadanego czasu, nawet po spadku obiektu, jednak wykres uwzględnia tylko część trajektorii przed spadkiem. Ustawienie parametru ```n``` na ```0``` spowoduje automatyczny dobór odpowiedniego parametru przez funkcję ```solve_ivp```.

````python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Stałe fizyczne
R_earth = 6.371e6      # Promień Ziemi (m)
MG = 3.986e14          # Stała grawitacji razy masa Ziemi (m^3/s^2)

#parametry do symulacji
drag_coef = 0.47               # Współczynnik oporu aerodynamicznego (dla kuli jest stały), wartość z ze strony 'Drag coefficient' na Wikipedii
cross_section = 0.1            # Przekrój poprzeczny (m^2)
mass = 1                       # Masa obiektu (kg)
h_start = 500_000              # Wysokość początkowa (m)
v_horizontal = 7_000           # Początkowa prędkość (m/s)
v_vertical = 0                 # Początkowa prędkość wzdłuż osi pionowej (m/s)
time = 2000                    # Czas symulacji w sekundach
n = 200                        # Próbkowanie czasu

# Funkcja gęstości atmosfery, obliczana dla wysokości nad poziomem morza, a nie odległości od środka Ziemii.
rho = lambda h:  1.225 * np.exp(-h/ 8500)  # gęstość w kg/m^3, h w metrach

def simulate_trajectory(initial_height, initial_velocity, C, A, m, duration, N=0, make_picture=False, fall_info=False):
    CA = C*A

    # Definicja układu równań różniczkowych
    def equations(t, y):
        # Rozpakowanie wektora stanu
        x, y, vx, vy = y

        # Obliczenie wysokości nad powierzchnią Ziemi
        r = np.sqrt(x**2 + y**2)
        h = r - R_earth

        # Gęstość atmosfery na danej wysokości
        if h > 0:  # Atmosfera istnieje tylko powyżej powierzchni Ziemi
            rho_h = rho(h)
        else:
            rho_h = 1.225  # Gęstość atmosfery na powierzchni Ziemi

        # Prędkość
        v = np.sqrt(vx**2 + vy**2)

        # Równania ruchu, te same, ale warto zauważyć, że rho_h jest wyliczane od wysokości
        fx = vx
        fy = vy
        ax = - (CA / (2 * m)) * rho_h * v * vx - (MG / r**3) * x
        ay = - (CA / (2 * m)) * rho_h * v * vy - (MG / r**3) * y

        return [fx, fy, ax, ay]

    # Parametry początkowe
    initial_position = [0, R_earth + initial_height]  # Początkowe położenie (x=0, y=100 km nad powierzchnią Ziemi)
    initial_position.extend(initial_velocity)
    y0 = initial_position
    # Przedział czasowy rozwiązania
    t_span = (0, duration)  # Czas w sekundach


    # Rozwiązanie równania
    if N == 0:
        if fall_info:
            solution = solve_ivp(equations, t_span, y0, rtol=1e-8, atol=1e-10, method='BDF', dense_output=True)
        else:
            solution = solve_ivp(equations, t_span, y0, rtol=1e-8, atol=1e-10, method='BDF')
    else: # Przypadek dla metody symplektycznej
        t_eval = np.linspace(*t_span, N)
        if fall_info:
            solution = solve_ivp(equations, t_span, y0, rtol=1e-8, atol=1e-10, t_eval=t_eval, method='BDF', dense_output=True)
        else:
            solution = solve_ivp(equations, t_span, y0, rtol=1e-8, atol=1e-10, t_eval=t_eval, method='BDF')

    # Filtracja punktów przed kolizją z Ziemią
    distances = np.sqrt(solution.y[0]**2 + solution.y[1]**2)
    valid_indices = distances > R_earth
    x_valid = solution.y[0][valid_indices]
    y_valid = solution.y[1][valid_indices]

    # Obliczenie maksymalnej odległości od środka Ziemi
    max_distance = max(np.max(distances), R_earth)

    # Sprawdzenie, czy obiekt spadł
    fall = len(x_valid) < len(solution.y[0])

    if fall:
        last_time_above_earth = solution.t[valid_indices][-1]  # Ostatni moment kiedy obiekt jest nad Ziemią
    else:
        last_time_above_earth = solution.t[-1]  # Ostatni moment symulacji jeśli obiekt nie spadł

    # Liczenie liczby okrążeń
    crossings = np.where((x_valid[:-1] < 0) & (x_valid[1:] >= 0))[0]
    num_orbits = len(crossings)
    #Robienie rysunku:
    if make_picture:
        # Wyświetlenie wyników
        if fall:

            print("Obiekt spadł na Ziemię.")
            print(f"Liczba pełnych okrążeń Ziemi przed spadkiem: {num_orbits}")

            #skalowanie od rozmiaru dancych
            if int(last_time_above_earth) < 500:
              print(f"Czas trwania spadku: {int(last_time_above_earth)} sekund")
              time_size = 's'
            elif int(last_time_above_earth) < 3600:
              print(f"Czas trwania spadku: {int(last_time_above_earth/60)} minut")
              time_size = 'm'
            elif int(last_time_above_earth) < 86400:
              print(f"Czas trwania spadku: {int(last_time_above_earth/3600)} godzin")
              time_size = 'h'
            elif int(last_time_above_earth) < 31556926:
              print(f"Czas trwania spadku: {int(last_time_above_earth/86400)} dni")
              time_size = 'd'
            else:
              print(f"Czas trwania spadku: {int(last_time_above_earth/31556926)} lat")
              time_size = 'y'

        else:
            print("Obiekt nie spadł na Ziemię.")
            print(f"Liczba pełnych okrążeń Ziemi w trakcie symulacji: {num_orbits}")
            time_size = 'y'

        # info o upadku (bądź jego braku)
        if fall_info:
            tsamples_of_fall = np.linspace(0, last_time_above_earth, 100)
            rs_during_fall = np.sqrt(np.sum(solution.sol(tsamples_of_fall)**2, axis=0))

            #skalowanie w zależności od rozmiaru danych
            if time_size == 'm':
              tsamples_of_fall = tsamples_of_fall / 60
            elif time_size == 'h':
              tsamples_of_fall = tsamples_of_fall / 3600
            elif time_size == 'd':
              tsamples_of_fall = tsamples_of_fall / 86400
            else:
              tsamples_of_fall = tsamples_of_fall / 31556926

            plt.figure(figsize=(8, 8))
            plt.plot(tsamples_of_fall, (rs_during_fall - R_earth) / 1000, label="Wysokość podczas spadania", color="black")
            plt.xlabel(f"t ({time_size})")
            plt.ylabel("h (km)")
            plt.title("Trajektoria obiektu spadającego na Ziemię")

        # Wykres trajektorii z Ziemią
        plt.figure(figsize=(8, 8))

        # Dodanie okręgu reprezentującego Ziemię
        earth = plt.Circle((0, 0), R_earth, color='blue', label="Ziemia", alpha=0.5)
        plt.gca().add_artist(earth)

        # Trajektoria obiektu przed kolizją z Ziemią
        plt.plot(x_valid, y_valid, label="Trajektoria", color="orange")

        # Oznaczenia osi i tytuł
        plt.xlabel("x [m]")
        plt.ylabel("y [m]")
        plt.title("Trajektoria obiektu spadającego na Ziemię")

        # Dostosowanie widoku
        plt.grid()
        plt.axis('equal')  # Zachowanie proporcji
        plt.legend()
        plt.xlim(-1.1 * max_distance, 1.1 * max_distance)
        plt.ylim(-1.1 * max_distance, 1.1 * max_distance)
        plt.show()

    return [fall, num_orbits, max_distance, last_time_above_earth]

simulate_trajectory(1_000_000, [3000, 0], 0.47, 10, 500, 10_000_000, 0, make_picture=True, fall_info=False)
````

##Kod testu na stabliność orbit

````python

# wszystkie stałe
heights = np.array([1, 5, 10, 50, 100, 120, 140, 150, 160, 165, 170, 173, 177, 180, 183, 186, 190, 192, 194, 196, 198, 200, 202, 204, 206, 208, 210, 215, 220, 225, 230, 240]) * 1000  # wysokości początkowe w m
MG = 3.986e14 #stała m^3/s^2
R_earth = 6.371e6 #w metrach
drag_coef = 0.1
cross_section = 5
time = 10500000   # miesiąc w sekundach
n = 0
mass = 500

Rs = heights + R_earth

results = []
stable_velocities = np.sqrt(MG * np.ones(len(heights)) / Rs ) # wzór z wyżej

for h, v in zip(heights, stable_velocities):
    # print(f"Obliczam czas spadnia z wyokości {h // 1000} km z prędkością początkową {round(v / 1000, 2)} km/h.")
    if h == 400000:
      result = simulate_trajectory(h, [v, 0], drag_coef, cross_section, mass, time, n, True)
    else:
      result = simulate_trajectory(h, [v, 0], drag_coef, cross_section, mass, time, n, False)
    results.append(result)

results = np.array(results)

plt.figure(figsize=(8, 6))
plt.plot(heights/1000, results[:,3] // (24*3600), marker='.', linestyle='-', color='k')
plt.xlabel('Wysokość n.p.m (km)')
plt.ylabel('Czas orbitowania, t_max = {} dni'.format(time // (24*3600)))
plt.grid(True)
plt.show()
````

###Kod do jednoczesnego symulowania trajektorii wielu odłamków
W celu powtórzenia testu poniżej należy odpalić tę komórkę.

````python
#Startuje z tego samego punktu (0, initial_height). Z prędkością [vel[0], vel[1]]
#Na wejściu powinien znajdować się wektor z parametrami osobno dla każdego elementu,
#na przykład dla dwóch obiektów wektor wysokości początkowych to [h1, h2], a wektor prędkości początkowych [[v1x, v1y], [v2x, v2y]]
#
#Oferuje przyspieszenie jedynie podczas symulowania trajektorii kilku niespadających obiektów, gdyż w przeciwnym przypadku
#adaptacyjne metody zawarte w implementacji solve_ivp szybciej wykrywają upadek.
def satellite_dynamics(duration, initial_heights : np.ndarray, initial_velocities : np.ndarray, cross_sections : np.ndarray, masses : np.ndarray, printt=False):
    CAs = cross_sections * C
    R_earth = 6.371e6
    MG = 3.986e14
    y0 = np.array([[0, hght + R_earth, vel, 0] for hght, vel in zip(initial_heights, initial_velocities)]).flatten()
    def equations(t, state: np.ndarray):
        x = state[::4]  # Pozycje x
        y = state[1::4]  # Pozycje y
        vx = state[2::4] # Prędkości x
        vy = state[3::4] # Prędkości y
        r = np.sqrt(x**2 + y**2)
        h = r - R_earth
        rho_h = np.piecewise(h, [h > 0, h <= 0], [lambda h: 1.225 * np.exp(-h / 8000), 100000])
        v = np.sqrt(vx**2 + vy**2)
        ax = - (CAs/ (2 * masses)) * rho_h * v * vx - (MG / r**3) * x
        ay = - (CAs / (2 * masses)) * rho_h * v * vy - (MG / r**3) * y
        return np.column_stack((vx, vy, ax, ay)).ravel()

    solution = solve_ivp(equations, (0, duration), y0, rtol=1e-8, atol=1e-10, method='BDF')
    distances = []
    k = 0
    for i in range(len(solution.y[:,0])//4) :
        if any(np.sqrt(solution.y[i*4,:]**2 + solution.y[i*4 + 1,:]**2) < R_earth):
            k = k + 1



    if printt:
        plt.figure(figsize=(8, 8))

        # Dodanie okręgu reprezentującego Ziemię
        earth = plt.Circle((0, 0), R_earth, color='blue', label="Ziemia", alpha=0.5)
        plt.gca().add_artist(earth)
        max_distance = np.max(np.linalg.norm(solution.y[:2, :], axis=0))

        # Trajektoria obiektu przed kolizją z Ziemią
        for i in range(len(solution.y[:,0])//4) :
            plt.plot(solution.y[i*4,:], solution.y[i*4 + 1,:], color="orange")

        # Oznaczenia osi i tytuł
        plt.xlabel("x [m]")
        plt.ylabel("y [m]")
        plt.title("Trajektoria obiektu spadającego na Ziemię")

        # Dostosowanie widoku
        plt.grid()
        plt.axis('equal')  # Zachowanie proporcji
        plt.legend()
        plt.xlim(-1.1 * max_distance, 1.1 * max_distance)
        plt.ylim(-1.1 * max_distance, 1.1 * max_distance)
        plt.show()
    result = round(k/(len(solution.y[:,0])//4)*100, 2)
    return result

C = 2.2
satellite_dynamics(10000, np.array([10000]), np.array([200]), np.array([5]), np.array([100]), printt=True)
````

# 3. Model zderzenia - metody Monte Carlo

## Wyprowadzenie i opis metody

---
Zderzenia, w których satelita ulega rozpadowi a jego fragmenty zachowują się chaotycznie po kolizji mogą być dodatkowym impulsem ułatwiającym spadek masy wyniesionej na orbitę, nawet jeśli przed zderzeniem jej orbita była stabilna.
W dalszej części projektu chcemy oszacować wartość oczekiwaną masy obu satelit ulegającą deorbitacji po ich centralnym (dość wyidealizowanym) zderzeniu, zakładając dla uproszczenia modelu, równe masy satelit. Analizując zachowanie odłamków satelity po rozpadzie zakładamy, że wektory prędkości po rozpadzie satelity pochodzą z rozkładu normalnego i są niezależne od siebie. Ponadto, przy założeniu identycznych mas i przeciwnych prędkości satelit na podstawie zasady zachowania pędu oczekujemy od tych rozkładów aby
$$ 0 = \sum_{i=1,2}M\overrightarrow{v_i}_{przed \ zderzeniem} = \mathbb{E}\sum_im_i\vec{v_i} = \mathbb{E}\vec{v_1} \sum_im_i=M\mathbb{E}\vec{v_1}$$
$$\mathbb{E}\vec{v_i} = 0$$
zatem oczekujemy żeby wektory prędkości pochodziły z rozkładu o wartości oczekiwanej 0.
Wyznaczmy wariancję z zasady zachowania energii
$$Var(v_i) = \mathbb{E}v_i^2 - (\mathbb{E}v_i)^2 = \mathbb{E}v_i^2 - 0 = \mathbb{E}v_i^2$$
$$ \sum_{i=1,2}\frac{M{v_i}_{przed \ zderzeniem}^2}{2} = \mathbb{E}\sum_i \frac{m_iv_i^2}{2} =  \left( \sum\frac{m_i}{2} \right) \left(\mathbb{E}v_1^2\right) = \frac{M}{2}\mathbb{E}v_i^2$$
$$ Var(v_i) = \mathbb{E}v_i^2 = v_{przed}^2 $$
Jako że nie mamy do czynienia ze zderzeniem sprężystym, w którym zachowana jest cała energia kinetyczna, zatem powinniśmy uwzględnić stałą $C_z$ oznaczającą część energii utraconej w trakcie zderzenia. Niestety jednak ta wartość całkowicie zależy od przebiegu zderzenia i nie potrafimy jej oszacować, niemniej, chcąc oszacować wartość oczekiwaną masy satelity, która spadnie na ziemię, sytuacja, w której $C_z=1$ jest dla nas najgorszym przypadkiem. Istotnie, im większa wariancja tym większe wartości prędkości odłamków, i mniejsza szansa, że spowolnią na tyle aby spaść z orbity. W szczególności dla $C_z=0$ (albo bardzo blisko $0$) mamy do czynienia ze zderzeniem idealnie niesprężystym gdzie satelity łączą się ze sobą znosząc wzajemnie swój pęd i spadają w całości na ziemię.

Zatem rozsądnie będzie przyjąć do naszego modelu rozkład wektorów prędkości jako
$$ v_i \sim \mathcal{N}(0,v_{przed}^2 Id)$$
natomiast w symulacjach uwarunkować przy losowaniu próbki po zderzeniu, w których łączna sumia energii kinetycznych jest zachowana.

Dodatkowo chcemy w prosty sposób przybliżyć liczbę odłamków powstałych w wyniku rozpadu satelity oraz rozkład ich mas w celu oszacowania wartości oczekiwanej masy ulegającej deorbitacji metodą Monte Carlo.
W tym celu wykorzystamy rozkład Poissona (do wyznaczenia ilości fragmentów $n$) oraz rozkład Dirichleta (odpowiadających rozkładowi długości fragmentów odcinka jednostkowego podzielonego jednostajnie na $n$ części, gdzie utożsamiamy część masy całkowitej jaką stanowi $i$-ty odłamek z długością $i$-tego fragmentu odcinka)
Zatem
$$ n \sim Poiss (k)$$
$$ (m_1,m_2,...,m_n) \sim Dir(1,1...,1)$$


##Analiza modelu metodą Monte Carlo


---
W dalszej symulacji losujemy rozpad satelity zgodnie z wyżej opisanym rozkładem, następnie korzystając z wcześniej napisanego układu równań opisującego trajektorię sprawdzamy czy dany odłamek spadł na ziemię, czy nie, i w ten sposób wyznaczamy procentową część masy satelity i powstałych odłamków, które uległy deorbitacji po upłynięciu 1 roku.


Przeprowadzając małą liczbę symulacji heurystycznie oszacowaliśmy wariancję otrzymanego estymatora Monte Carlo, a następnie korzystając z tej wartości wyznaczyliśmy liczbę prób, którą musimy wykonać aby uzyskać wstępną żądaną dokładność +-1 punktu procentowego z prawdopodobieństwem 95% na $\leq200$ powtórzeń.(W naszych warunkach obliczenia związane z pojedynczą trajektorią w rocznym horyzoncie czasowym trwały około 20s, co przy dużej liczbie prób, odłamków i wysokości generuje długi czas trwania całej symulacji - około 11 godzin). ## Kod funkcji symulującej rozpad obiektu

````python
import numpy as np
def boom(v,m,k):
    #  rozbija satelitę na drobny mak
    #  k: przyjmuje parametr k do rozkładu Poissona (wartość oczekiwana liczby odłamków) - na ten moment musimy to po prostu uznać za prametr modelu
    #  v: sumę wartości prędkości satelit przed zderzeniem
    #  tzn. dla satelit zderzających się z prędkością kosmiczną v0 chcemy aby v = 2 * v0 WAŻNE!
    #  m: masę satelity

    #  zwraca: listę (losowej długości) kawałków satelity opisaną listami z parametrami
    #  [[v1,v2],A,m]
    #  gdzie [v1,v2] to wektor prędkości fragmentu (poziomy/pionowy wymiennie), m to jego masa a A to nowy przekrój
    n = np.random.poisson(lam=k)+1 #chcemy żeby był przynajmniej jeden obiekt do śledzenia
    vel = np.random.normal(loc=0, scale = v^2, size = (n,2))
    Dir = np.random.uniform(size = n-1)
    Dir = np.append(Dir,[0,1])
    Dir.sort()
    mass = []
    cross = []  #znając proporcje masy możemy od razu wyznaczyć przekrój nowej kuli (odłamka)
    out = []
    for i in range(n): mass.append(m*(Dir[i+1]-Dir[i]))
    for i in range(n): cross.append((mass[i] / 37.97) ** (100 / 186))
    for i in range(n): out.append((vel[i], cross[i], mass[i]))
    return out


def fall_analysis(data, time, height):
  #   data: tupla z funkcji boom
  #   time: horyzont czasowy zadania
  tot_obj=len(data)
  count_el=0
  count_m=0
  total_m=0

  for element in data:
    v, A, m = element
    C=1
    result = simulate_trajectory(height, v, C, A, m, time, 0, make_picture=False, fall_info=False)
    if result[0]:
      count_el += 1
      count_m += m
    total_m += m

  return[100*count_el/tot_obj, 100*count_m/total_m ] #zwraca procent obiektów które spadły i procent masy która spadła
````

##Kod programu wykonującego symulację Monte Carlo

````python
import time as timer
def MC(h):
    #wysokość w METRACH!
    m = 350                      #masa - masa satelity jest dużo większa ale symulujemy mniej fragmentów
    k = 150                     #średnia ilość odłamków
    time = 32_000_000           #rok w sekundach
    #time = 3            #czas testowy
    v = float(stable_velocity(h))    #prędkość kosmiczna
    reps = 200                  #powtórzenia do MC
    output = []
    for i in range(reps):
        data = boom(v,m,k)
        output.append(fall_analysis(data,time,h))

    return output


heights = np.linspace(200,2000,10)
variances = []
averages =[]
start = timer.time()
for h in heights:
    name = str(h) + ".txt"
    h=h*1000    #konwersja km -> m
    result = MC(h)
    variances.append([np.var(result[0]), np.var(result[1])])
    averages.append([np.mean(result[0]), np.mean(result[1])])
    with open(name, "w") as file:
        for i in result:
            file.write(str(i[0])+ "\t" + str(i[1])+"\n")

with open("variances.txt", "w") as file:
        for i in variances:
            file.write(str(i[0])+ "\t" + str(i[1])+"\n")
with open("averages.txt", "w") as file:
        for i in averages:
            file.write(str(i[0])+ "\t" + str(i[1])+"\n")

print("done")
print("time: ",timer.time()-start)
````

Przeprowadzając symulację stochastyczną otrzymaliśmy główny wynik wieńczący ten fragment projektu.
Zgodnie z wynikami symulacji, w konsekwencji tak przebiegającego rozpadu satelit na niskich orbitach okołoziemskich, w przeciągu roku deorbitacji ulega około 40% masy ulegającej całkowitemu rozpadowi satelity. Ponadto ta wartość zależy od wysokości, na której przebiega zderzenie - maleje ona od 50% na wysokości 200 km nad powierzchnią ziemi do około 25% dla wyższych orbit zbliżających się do 2000 km.


Warto zwrócić uwagę, że ten sposób modelowania zachowania odłamków satelit może okazać się szczególnie wartościowy przy badaniu zachowania najmniejszych odłamków powstających w wyniku eksplozji, które zachowują się w mniej przewidywalny sposób i są trudniejsze lub nawet niemożliwe w wykrywaniu i badaniu metodami empirycznymi, które natomiast dalej stanowią istotne zagrożenie.


Interpretując wyniki symulacji na wykresie powyżej należy również pamiętać o dość wyidealizowanych założeniach, które zostały przyjęte do konstrukcji modelu. Podczas prawdziwego zderzenia energia kinetyczna jest rozpraszana na inne rodzaje energii (jak ciepło, czy chociażby energia potrzebna do rozbicia satelity na fragmenty), natomiast nasz model pesymistycznie zakłada zachowanie energii kinetycznej odłamków. Przez to odłamki w naszej symulacji mają większą energię kinetyczną co przekłada się na większą szansę pozostania na orbicie, sugerując, że wyniki te będą dolnym oszacowaniem części masy, która spada z powrotem na ziemię.


Oczywiście program do przeprowadzania symulacji nadaje się do dalszego użytku i badania zależności, których nie sprawdziliśmy jak i do poprawienia naszych wyników, rozwijając badanie problemu. Można na przykład:
- Otrzymać większą dokładność i pewność estymatora Monte Carlo zwiększając ilość powtórzeń, co jednak bezpośrednio wiąże się z wydłużeniem czasu obliczeń albo wymaga dostępu do większej mocy obliczeniowej
- Zbadać zachowanie odłamków w dłuższej perspektywie czasowej, jednakże należy wtedy zwrócić uwagę na nawarstwiający się błąd numeryczny, który może istotnie wpłynąć na poprawność wyniku
- Dokładniej wyznaczyć rodzaj zależności wartości oczekiwanej masy opadającej na ziemię z wysokością orbity, na której dochodzi do zderzenia, poprzez istotne zwiększenie gęstości siatki wysokości do przeprowadzonej symulacji.
